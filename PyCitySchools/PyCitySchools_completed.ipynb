```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_path = "Resources/schools_complete.csv"
student_data_path = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_path)
student_data = pd.read_csv(student_data_path)

# Combine the data into a single dataset.  
school_combined = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_combined.head()
```

```python
##Create a high-level snapshot of the district's key metrics in a DataFrame

#Find total number of schools
total_schools = school_combined['school_name'].value_counts().count()
#Find total number of students
total_students = school_combined['student_name'].count()
#Find total of budgets
total_budget = school_combined['budget'].unique().sum()
#Find average math scores
average_math = school_combined['math_score'].mean()
#Find average reading score
average_read = school_combined['reading_score'].mean()
#Find % of students passing math
math_pass = school_combined.loc[school_combined['math_score'] >= 70,('math_score')].count() / school_combined['math_score'].count() * 100
#Find % of students passing reading
read_pass = school_combined.loc[school_combined['reading_score'] >= 70,('reading_score')].count() / school_combined['reading_score'].count() * 100
#Find Overall percent of students passing
math_read = school_combined.loc[(school_combined['math_score'] >= 70) & 
                                (school_combined['reading_score'] >= 70), ('math_score')].count()
overall_pass = math_read / school_combined['student_name'].count() * 100

```

```python
##Print out high level snapshot data frame
snapshot_df = pd.DataFrame ({
    "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget" : [total_budget],
    "Average Math Score" : [average_math],
    "Average Reading Score" : [average_read],
    "% Passing Math" : [math_pass],
    "% Passing Reading": [read_pass],
    "% Overall Passing" : [overall_pass]
})
snapshot_df
```

```python
##Create an overview table that summarizes key metrics about each school

#Find school type and name
school_type = school_data.set_index(["school_name"])["type"]

#Find Total Number of students
total_students = school_combined['school_name'].value_counts()

#Find Total School Budget
total_budget = school_combined.groupby(["school_name"])["budget"].first()

#Find per Student Budget
per_student_budget = total_budget / total_students

#Find Average math score
math_per_school = school_combined.groupby(["school_name"])["math_score"].mean()

#Find Average reading score
read_per_school = school_combined.groupby(["school_name"])["reading_score"].mean()

#Find % Passing math
students_passing_math = school_combined[school_combined["math_score"] >= 70].groupby(["school_name"]).count()["Student ID"]
percent_pass_math = (students_passing_math / total_students) * 100

#Find % passing reading
students_passing_reading = school_combined[school_combined["reading_score"] >= 70].groupby(["school_name"]).count()["Student ID"]
percent_pass_reading = (students_passing_reading / total_students) * 100

#Find overall Passing
students_passing_both = school_combined[(school_combined["math_score"] >= 70) & 
                                       (school_combined["reading_score"] >= 70)].groupby(["school_name"]).count()["Student ID"]
passing_both_percent = (students_passing_both / total_students) * 100
```

```python
##Create dataframe to hold above results
school_summary_df = pd.DataFrame({  
    "School Type": school_type, 
    "Total Students": total_students, 
    "Total School Budget": total_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": math_per_school,
    "Average Reading Score": read_per_school,
    "% Passing Math": percent_pass_math, 
    "% Passing Reading": percent_pass_reading,
    "% Overall Passing": passing_both_percent
})
school_summary_df
```

```python
##Sort and display the top five performing schools by % overall passing.
top_performing_schools = school_summary_df.sort_values(["% Overall Passing"], ascending = False).head()
top_performing_schools
```

```python
##Sort and display the five worst-performing schools by % overall passing.
bottom_performing_schools = school_summary_df.sort_values(["% Overall Passing"]).head()
bottom_performing_schools
```

```python
##Create a table that lists the average math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

#Find ninth grade average
ninth_grade = student_data[student_data["grade"] == "9th"].groupby(["school_name"]).mean()["math_score"]

#Find tenth grade average
tenth_grade = student_data[student_data["grade"] == "10th"].groupby(["school_name"]).mean()["math_score"]

#Find eleventh grade average
eleveneth_grade = student_data[student_data["grade"] == "11th"].groupby(["school_name"]).mean()["math_score"]

#Find twelvth grade average
twelvth_grade = student_data[student_data["grade"] == "12th"].groupby(["school_name"]).mean()["math_score"]

#create dataframe
math_scores_by_grade = pd.DataFrame({   
    "9th Grade": ninth_grade.map("{:.1f}".format),
    "10th Grade": tenth_grade.map("{:.1f}".format),
    "11th Grade": eleveneth_grade.map("{:.1f}".format),
    "12th Grade": twelvth_grade.map("{:.1f}".format)
})

math_scores_by_grade
```

```python
##Create a table that lists the average reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

#Find ninth grade reading average
ninth_grade_read = student_data[student_data["grade"] == "9th"].groupby(["school_name"]).mean()["reading_score"]

#Find tenth grade reading average
tenth_grade_read = student_data[student_data["grade"] == "10th"].groupby(["school_name"]).mean()["reading_score"]

#Find eleventh grade reading average
eleveneth_grade_read = student_data[student_data["grade"] == "11th"].groupby(["school_name"]).mean()["reading_score"]

#Find twelvth grade reading average
twelvth_grade_read = student_data[student_data["grade"] == "12th"].groupby(["school_name"]).mean()["reading_score"]

#create dataframe
reading_scores_by_grade = pd.DataFrame({   
    "9th Grade": ninth_grade_read.map("{:.1f}".format),
    "10th Grade": tenth_grade_read.map("{:.1f}".format),
    "11th Grade": eleveneth_grade_read.map("{:.1f}".format),
    "12th Grade": twelvth_grade_read.map("{:.1f}".format)
})

reading_scores_by_grade
```

```python
##Create a table that breaks down school performances based on average Spending Ranges (Per Student)

# Create bins
bins = [0, 585, 630, 645, 680]
group_name = ["<$585", "$585-630", "$630-645", "$645-680"]

#Add ranges to dataframe
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=group_name)

#group dataframe by ranges created
group_by_spending = school_summary_df.groupby(["Spending Ranges (Per Student)"])

#get values to creat new dataframe
grouped_spending_sum = grouped_spending.mean()


#create
score_by_spending = grouped_spending_sum[["Average Math Score",
                                    "Average Reading Score",
                                    "% Passing Math",
                                    "% Passing Reading",
                                    "% Overall Passing"]]
score_by_spending

```

```python
##Create a table that breaks down school performances based on school size

# Create bins
bins = [0, 1000, 2000, 5000]
group_name = ["Small <1000", "Medium (1000-2000)", "Large (2000-5000)"]

#Add ranges to dataframe
school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], bins, labels=group_name)

#group dataframe by ranges created
group_by_size = school_summary_df.groupby(["School Size"])

#get values to creat new dataframe
grouped_size_sum = group_by_size.mean()


#create DataFrame
score_by_size = grouped_size_sum[["Average Math Score",
                                    "Average Reading Score",
                                    "% Passing Math",
                                    "% Passing Reading",
                                    "% Overall Passing"]]
score_by_size
```

```python
##Create a table that breaks down school performances based on school type

#group by school type
group_by_type = school_summary_df.groupby(["School Type"])

#get values to creat new dataframe
grouped_type_sum = group_by_type.mean()


#create DataFrame
score_by_type = grouped_type_sum[["Average Math Score",
                                    "Average Reading Score",
                                    "% Passing Math",
                                    "% Passing Reading",
                                    "% Overall Passing"]]
score_by_type
```
